```python
import openai 

openai.api_key = "sk-API_KEY"

def ask_GPT(query): 
	completion = openai.ChatCompletion.create( 
		model="gpt-3.5-turbo", 
		temperature=0.8, 
		max_tokens=2000, 
		messages=[ 
			{"role": "user", "content": query} 
		] 
	) 
	return completion.choices[0].message.content 

while True: 
	query = input() 
	if query == "exit": 
		print("Program Terminated") 
		break
	response = ask_GPT(query) 
	print(response)
```
